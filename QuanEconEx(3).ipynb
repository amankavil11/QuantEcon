```python
#Exercise 3.6.1 for QuantEcon
import numpy as np
import matplotlib.pyplot as plt
T = 200
a = 0.9

x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = a * x[t] + np.random.randn()
#default x-axis scale of plot is len x[t]
plt.plot(x)
plt.show
    
   
```

```python
#Exercise 3.6.2 for QuantEcon

a = np.array([0.0, 0.8, 0.98])
T =200
x = np.empty(T+1)
x[0] = 0

for alpha in a:
    for t in range(T):
        x[t+1] = alpha * x[t] + np.random.randn()
    plt.plot(x, label= "alpha = " + str(alpha))
plt.legend()




```

```python
#Exercise 3.6.3 for QuantEcon
for t in range(T):
    x[t+1] = a * np.abs(x[t]) + np.random.randn()
plt.plot(x)
plt.show
```

```python
#Exercise 3.6.4 for QuantEcon
for t in range(T):
    if(x[t] >= 0):
        x[t+1] = a * x[t] + np.random.randn()
    else:
        x[t+1] = a * -x[t] + np.random.randn()
plt.plot(x)
plt.show
```

```python
#Exercise 3.6.5 for QuantEcon

#the higher the sample count, 'n', the more accurate "PI" value is, naturally
n = 1000000

count = 0
for i in range(n):
    #np.random.uniform() has default floor value of 0 and default ceiling value of 1
    u, v = np.random.uniform(), np.random.uniform()
    #distance formula/circle equation to determine points within the circle
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d <= 0.5:
        count += 1
#ratio of inside to circle to total points, which is proportional to the circles area and the squares area
area_estimate = count / n
#Area of rectangle = (2r)^2 = 4r^2
#Area of circle = PI*r^2
#PI*r^2/4r^2 = PI/4
Pie = area_estimate * 4
print(Pie)
```
