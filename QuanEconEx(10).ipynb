```python
#Exercise 10.5.1 for QuantEcon
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

def f(x, theta):
    return np.cos(np.pi*theta*x) * np.exp(-x)

x = np.linspace(0,5,1000)
theta_vals =  np.linspace(0,2,10)

for theta in theta_vals:
    ax.plot(x, f(x, theta))
```
